```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq


# Read JSON files into pandas dataframes
tpcc_25 = pd.read_json('/home/habib/Downloads/equation_train_data/real/TPCC-OP-25-S-1.25/W2K8.TPCC.10-19-2007.12-51-PM.trace-filtered-target-converted-split.txt (1).json')
tpcc_10 = pd.read_json('/home/habib/Downloads/equation_train_data/real/TPCC-OP-25-S-1.25/W2K8.TPCC.10-19-2007.12-51-PM.trace-filtered-target-converted-split.txt (2).json')
prxy_25 = pd.read_json('/home/habib/Downloads/equation_train_data/real/msr_prxy_25/msr_cambridge_1.txt.json')
prxy_10 = pd.read_json('/home/habib/Downloads/equation_train_data/real/msr_prxy_25/msr_cambridge_1.txt (1).json')
```

```python
tpcc_25_invalid_page_count = tpcc_25.iloc[4].reset_index(drop=True).to_frame(name='invalid_pages')
tpcc_10_invalid_page_count = tpcc_10.iloc[4].reset_index(drop=True).to_frame(name='invalid_pages')
prxy_25_invalid_page_count = prxy_25.iloc[4].reset_index(drop=True).to_frame(name='invalid_pages')
prxy_10_invalid_page_count = prxy_10.iloc[4].reset_index(drop=True).to_frame(name='invalid_pages')
```

```python
N = len(tpcc_25_invalid_page_count)
T = 1.0

tpcc_10_yf = fft(tpcc_25_invalid_page_count)
tpcc_25_yf = fft(tpcc_10_invalid_page_count)
prxy_10_yf = fft(prxy_25_invalid_page_count)
prxy_25_yf = fft(prxy_10_invalid_page_count)

xf = fftfreq(N, T)[:N//2]

plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(20, 10))
plt.plot(xf, 2.0/N * np.abs(prxy_10_yf[0:N//2]), label='msr_prxy_10')
plt.plot(xf, 2.0/N * np.abs(tpcc_10_yf[0:N//2]), label='TPCC-OP-10-S-1.25')
plt.plot(xf, 2.0/N * np.abs(tpcc_25_yf[0:N//2]), label='TPCC-OP-25-S-1.25')
plt.plot(xf, 2.0/N * np.abs(prxy_25_yf[0:N//2]), label='msr_prxy_25')


plt.xlabel('Frequency')
plt.ylabel('Amplitude')

plt.legend()
plt.show()

```

```python
tpcc_25_mean = np.mean(tpcc_25_invalid_page_count['invalid_pages'])
tpcc_25_std_dev = np.std(tpcc_25_invalid_page_count['invalid_pages'])
tpcc_10_mean = np.mean(tpcc_10_invalid_page_count['invalid_pages'])
tpcc_10_std_dev = np.std(tpcc_10_invalid_page_count['invalid_pages'])
prxy_25_mean = np.mean(prxy_25_invalid_page_count['invalid_pages'])
prxy_25_std_dev = np.std(prxy_25_invalid_page_count['invalid_pages'])
prxy_10_mean = np.mean(prxy_10_invalid_page_count['invalid_pages'])
prxy_10_std_dev = np.std(prxy_10_invalid_page_count['invalid_pages'])
```

```python
print(f"TPCC-OP-25-S-1.25: mean = {tpcc_25_mean}, std_dev = {tpcc_25_std_dev}")
print(f"TPCC-OP-10-S-1.25: mean = {tpcc_10_mean}, std_dev = {tpcc_10_std_dev}")
print(f"msr_prxy_25: mean = {prxy_25_mean}, std_dev = {prxy_25_std_dev}")
print(f"msr_prxy_10: mean = {prxy_10_mean}, std_dev = {prxy_10_std_dev}")
```

```python
# create a cumulative array of invalid pages
tpcc_25_invalid_page_count = tpcc_25_invalid_page_count.cumsum()
tpcc_10_invalid_page_count = tpcc_10_invalid_page_count.cumsum()
prxy_25_invalid_page_count = prxy_25_invalid_page_count.cumsum()
prxy_10_invalid_page_count = prxy_10_invalid_page_count.cumsum()

# create a time array
time = np.arange(0, len(tpcc_25_invalid_page_count), 1)

# plot the data
fig, ax = plt.subplots()
ax.plot(time, tpcc_25_invalid_page_count, label='TPCC-OP-25-S-1.25')
ax.plot(time, tpcc_10_invalid_page_count, label='TPCC-OP-10-S-1.25')
ax.plot(time, prxy_25_invalid_page_count, label='msr_prxy_25')
ax.plot(time, prxy_10_invalid_page_count, label='msr_prxy_10')

ax.set(xlabel='Time', ylabel='Invalid Pages',
         title='Invalid Pages vs Time')
ax.grid()
ax.legend()

plt.show()


```

```python
def group_and_average(data, n=2):
    return np.array([np.mean(data[i:i + n]) for i in range(0, len(data), n)])

bin_size = 2
x_tpcc_25 = np.arange(0, len(tpcc_25_invalid_page_count), bin_size)
x_tpcc_10 = np.arange(0, len(tpcc_10_invalid_page_count), bin_size)
x_prxy_25 = np.arange(0, len(prxy_25_invalid_page_count), bin_size)
x_prxy_10 = np.arange(0, len(prxy_10_invalid_page_count), bin_size)

tpcc_25_binned = group_and_average(tpcc_25_invalid_page_count['invalid_pages'], bin_size)
tpcc_10_binned = group_and_average(tpcc_10_invalid_page_count['invalid_pages'], bin_size)
prxy_25_binned = group_and_average(prxy_25_invalid_page_count['invalid_pages'], bin_size)
prxy_10_binned = group_and_average(prxy_10_invalid_page_count['invalid_pages'], bin_size)


plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(20, 10))

plt.bar(x_tpcc_25, tpcc_25_binned, width=bin_size, color='b', label='TPCC-OP-25-S-1.25')
plt.bar(x_tpcc_10, tpcc_10_binned, width=bin_size, color='g', label='TPCC-OP-10-S-1.25')
plt.bar(x_prxy_25, prxy_25_binned, width=bin_size, color='r', label='msr_prxy_25')
plt.bar(x_prxy_10, prxy_10_binned, width=bin_size, color='y', label='msr_prxy_10')


plt.xlabel('Block')
plt.ylabel('Invalid Page Count')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq


# Read JSON files into pandas dataframes
tpcc_25 = pd.read_json('/home/habib/Downloads/equation_train_data/real/TPCC-OP-25-S-1.25/W2K8.TPCC.10-19-2007.12-51-PM.trace-filtered-target-converted-split.txt (1).json')
tpcc_10 = pd.read_json('/home/habib/Downloads/equation_train_data/real/TPCC-OP-25-S-1.25/W2K8.TPCC.10-19-2007.12-51-PM.trace-filtered-target-converted-split.txt (2).json')
prxy_25 = pd.read_json('/home/habib/Downloads/equation_train_data/real/msr_prxy_25/msr_cambridge_1.txt.json')
prxy_10 = pd.read_json('/home/habib/Downloads/equation_train_data/real/msr_prxy_25/msr_cambridge_1.txt (1).json')

tpcc_25_invalid_page_count = tpcc_25.iloc[4].reset_index(drop=True).to_frame(name='invalid_pages')
tpcc_10_invalid_page_count = tpcc_10.iloc[4].reset_index(drop=True).to_frame(name='invalid_pages')
prxy_25_invalid_page_count = prxy_25.iloc[4].reset_index(drop=True).to_frame(name='invalid_pages')
prxy_10_invalid_page_count = prxy_10.iloc[4].reset_index(drop=True).to_frame(name='invalid_pages')

tpcc_25_mean = np.mean(tpcc_25_invalid_page_count['invalid_pages'])
tpcc_25_std_dev = np.std(tpcc_25_invalid_page_count['invalid_pages'])
tpcc_10_mean = np.mean(tpcc_10_invalid_page_count['invalid_pages'])
tpcc_10_std_dev = np.std(tpcc_10_invalid_page_count['invalid_pages'])
prxy_25_mean = np.mean(prxy_25_invalid_page_count['invalid_pages'])
prxy_25_std_dev = np.std(prxy_25_invalid_page_count['invalid_pages'])
prxy_10_mean = np.mean(prxy_10_invalid_page_count['invalid_pages'])
prxy_10_std_dev = np.std(prxy_10_invalid_page_count['invalid_pages'])

# Compute cumulative standard deviation
def cumulative_std(data):
    return [np.std(data[:i]) if i > 1 else 0 for i in range(1, len(data) + 1)]

std_prxy_25 = cumulative_std(prxy_25_invalid_page_count['invalid_pages'])
std_tpcc_25 = cumulative_std(tpcc_25_invalid_page_count['invalid_pages'])
std_prxy_10 = cumulative_std(prxy_10_invalid_page_count['invalid_pages'])
std_tpcc_10 = cumulative_std(tpcc_10_invalid_page_count['invalid_pages'])

plt.rcParams.update({'font.size': 19})
plt.figure(figsize=(12, 6))
x_tpcc_25 = np.arange(0, len(tpcc_25_invalid_page_count), 1)
x_tpcc_10 = np.arange(0, len(tpcc_10_invalid_page_count), 1)
x_prxy_25 = np.arange(0, len(prxy_25_invalid_page_count), 1)
x_prxy_10 = np.arange(0, len(prxy_10_invalid_page_count), 1)

# plot a dot chart
plt.plot(x_prxy_25, prxy_25_invalid_page_count['invalid_pages'], 'ro', ms=0.4)
plt.plot(x_tpcc_25, tpcc_25_invalid_page_count['invalid_pages'], 'mo', ms=0.4)
plt.plot(x_prxy_10, prxy_10_invalid_page_count['invalid_pages'], 'bo', ms=0.4)
plt.plot(x_tpcc_10, tpcc_10_invalid_page_count['invalid_pages'], 'go', ms=0.4)

# Plot the cumulative standard deviation as a line
plt.plot(x_prxy_25, std_prxy_25, 'r-', linewidth=2, label='C.DoIPD prxy: OPS 25%')
plt.plot(x_tpcc_25, std_tpcc_25, 'm-', linewidth=2, label='C.DoIPD TPC-C: OPS 25%')
plt.plot(x_prxy_10, std_prxy_10, 'b-', linewidth=2, label='C.DoIPD prxy: OPS 10%')
plt.plot(x_tpcc_10, std_tpcc_10, 'g-', linewidth=2, label='C.DoIPD TPC-C: OPS 10%')

# # Add gray dots at the end points
# plt.scatter(x_prxy_25[-1], std_prxy_25[-1], color='gray', s=50, zorder=3)
# plt.scatter(x_tpcc_25[-1], std_tpcc_25[-1], color='gray', s=50, zorder=3)
# plt.scatter(x_prxy_10[-1], std_prxy_10[-1], color='gray', s=50, zorder=3)
# plt.scatter(x_tpcc_10[-1], std_tpcc_10[-1], color='gray', s=50, zorder=3)

y_offset = -1.5  # Adjust this value as needed
x_offset = -90  # Adjust this value as needed


plt.annotate('→', (x_prxy_25[-1] + x_offset, std_prxy_25[-1] + y_offset), fontsize=30, color='red', rotation=180+25)
plt.annotate('→', (x_tpcc_25[-1] + x_offset, std_tpcc_25[-1] + y_offset), fontsize=30, color='magenta', rotation=180+25)
plt.annotate('→', (x_prxy_10[-1] + x_offset, std_prxy_10[-1] - 1), fontsize=30, color='blue', rotation=180+25)
plt.annotate('→', (x_tpcc_10[-1] + x_offset, std_tpcc_10[-1] - 3), fontsize=30, color='green', rotation=180-25)

# # Add text next to each point
# plt.text(x_prxy_25[-1], std_prxy_25[-1], f'{std_prxy_25[-1]:.2f}', fontsize=18, verticalalignment='bottom', horizontalalignment='right')
# plt.text(x_tpcc_25[-1], std_tpcc_25[-1], f'{std_tpcc_25[-1]:.2f}', fontsize=18, verticalalignment='bottom', horizontalalignment='right')
# plt.text(x_prxy_10[-1], std_prxy_10[-1], f'{std_prxy_10[-1]:.2f}', fontsize=18, verticalalignment='bottom', horizontalalignment='right')
# plt.text(x_tpcc_10[-1], std_tpcc_10[-1], f'{std_tpcc_10[-1]:.2f}', fontsize=18, verticalalignment='bottom', horizontalalignment='right')


# draw a table on top of the figure 
cell_text = [['TPC-C', 25, round(tpcc_25_mean,2), round(tpcc_25_std_dev,2), 2.3],
             ['TPC-C', 10, round(tpcc_10_mean,2), round(tpcc_10_std_dev,2), 5.12],
             ['prxy', 25, round(prxy_25_mean,2), round(prxy_25_std_dev,2), 1.008],
             ['prxy', 10, round(prxy_10_mean,2), round(prxy_10_std_dev,2), 5.6]]
table = plt.table(cellText=cell_text, colLabels=['Workload', 'OPS (%)', 'Mean', 'DoIPD', 'WAF'], loc='top', cellLoc='center')
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1, 1.5)
for (i, j), cell in table.get_celld().items():
    if i == 0:
        cell.set_text_props(fontsize=19)
        cell.set_height(0.09)
    else:
        cell.set_height(0.09)
        cell.set_text_props(fontsize=18)
    
    
        

plt.xlabel('Block number')
plt.ylabel('IPC / C.DoIPD')
plt.legend(loc='upper right', bbox_to_anchor=(1, 0.85), fontsize=16, framealpha=0.3)
plt.show()

print(f"TPCC-OP-25-S-1.25: mean = {tpcc_25_mean}, std_dev = {tpcc_25_std_dev}")
print(f"TPCC-OP-10-S-1.25: mean = {tpcc_10_mean}, std_dev = {tpcc_10_std_dev}")
print(f"prxy_25: mean = {prxy_25_mean}, std_dev = {prxy_25_std_dev}")
print(f"prxy_10: mean = {prxy_10_mean}, std_dev = {prxy_10_std_dev}")
```
